```python
import pandas as pd
import numpy as np
import matplotlib as mp
import seaborn as sns
```

```python
train = pd.read_csv('zoo archive/zoo.csv')
train.head()
```

```python
classes = pd.read_csv('zoo archive/class.csv')
classes = classes.set_index('Class_Number')
classes.head()
```

```python
hashmap = {}
for index, row in classes.iterrows():
  animals = row['Animal_Names'].split(', ')
  for animal in animals:
    hashmap[animal] = index
hashmap
```

```python
train.describe()
```

```python
train.isna().sum()
```

```python
train.head()
```

```python
train.head()
```

```python
train['animal_name']= train['animal_name'].map(hashmap)
```

```python
train.head()
```

```python
y = train['class_type']
y
```

```python
train.drop('class_type',axis = 1,inplace = True)
train.head()
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(train, y)
```

```python
model = Perceptron()
```

```python
model.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, f1_score, precision_score
```

```python
predicted = model.predict(test_x)
```

```python
f1_score(predicted, test_y, average = 'micro')
```

```python
accuracy_score(predicted, test_y)
```

```python
precision_score(predicted, test_y, average = 'micro')
```

```python
recall_score(predicted, test_y, average = 'micro')
```

```python
confusion_matrix(predicted, test_y)
```
